```python
# import the packages we need to use
import pandas as pd
import numpy as np

# Load the data
data = pd.read_csv("/kaggle/input/pwc-task-3-data/Data.csv", sep=";")

# Display the first few rows of the data
data.head()
```

```python
# Rename the first column:
data = data.rename(columns={"Unnamed: 0": "Origination date"})
# Display the first few rows of the updated data 
data.head()
```

```python
# use the shape to find the number of vintages
print(data.shape)
vintage = data.shape[0]
```

```python
# get the table of cash flows that we already have
cash_flow = data.iloc[:, 2:]

# check the shape to make sure we get all columns and rows from original data
cash_flow.shape
cash_flow
```

```python
# get the amounts originated per vintage
amounts = np.array(data["Origination Amount"])
amounts
```

```python
# get the number of periods to be forecasted per vintage
forecast = 30 # All cash flows are assumed to be paid back over a period of 30 months
remain_periods = []
for i in reversed(range(vintage)):
    remain_periods.append(forecast - (i+1))
remain_periods = np.array(remain_periods)
remain_periods
```

```python
# calculate the repayment percentage for the cash flows
paid_percentage = cash_flow.div(amounts, axis = 0)
paid_percentage
```

```python
# get the original repayment percentage 
first_period = []
for i in range(vintage):
    first_period.append(paid_percentage.iloc[i , i])
first_period
```

```python
# based on the first period (original), get the second period
second_period = []

# The same holds true for the second month, except for the last vintage, December 2020, where the probability has not yet been observed
# because of the fact, the range should be (vintage - 1)
for i in range(vintage - 1):
    second_period.append(paid_percentage.iloc[i , i+1])

# For this vintage, assume that the expected repayment percentage for the second period will be twice the repayment of the first period
# because of the fact, we can add the last vintage to the list
second_period.append(first_period[-1] * 2)

second_period
```

```python
# create an empty table to store the expected repayment percentages
percentages = pd.DataFrame(np.zeros((vintage, forecast)))

# assign repayment percentage of first period and second period to the table
percentages.loc[:, 0] = first_period
percentages.loc[:, 1] = second_period

percentages
```

```python
# following the assumption to calculate the expected repayment percentages for the period from 3 to 30
for i in range(vintage):
    for j in range(2, forecast):
        # because in python, the range will start at 0, so based on the assumption function, we need minor change to fit in python
        percentages.iloc[i, j] = max(0, percentages.iloc[i, 1] * np.log(1 + (1 - np.sum(percentages.iloc[i, 0:j]))) * (1 - j  / forecast))

percentages

```

```python
# create a table of zeros for the forecasted expected repayment percentagess and starting in January 2021
p_forecast = pd.DataFrame(np.zeros((vintage, forecast - 1)))
# assign the expected repayment percentages to the correct periods
for i in range(vintage):
    for j in range(remain_periods[i]):
        p_forecast.iloc[i, j] = percentages.iloc[i, forecast - remain_periods[i] + j]

p_forecast
```

```python
# Assume that the forecasted cash flows are discounted with an annual rate of 2.5%.
discounted_rate = 0.025

# find the series of discount factors
discounted_factors = 1 / (1 + discounted_rate) ** (np.arange(1, forecast) / 12)
# present value of forecasted cash flows
present_value = np.transpose(np.transpose(p_forecast * discounted_factors) * amounts)

# set result is zero first
result = 0 
# use for loop to find the final sum of all present values
for i in present_value.sum():
    result += i
result
```

```python
# The client’s estimate of the portfolio value was CHF 84’993’122.67
# My own estimate is result 84779941.82
# Compute both the absolute and relative difference.
absolute_difference = 84993122.67 - round(result, 2)
absolute_difference = round(absolute_difference, 2)
relative_difference = absolute_difference / round(result, 2) * 100
relative_difference = str(round(relative_difference, 2)) +'%'

print(absolute_difference)
print(relative_difference)

# any difference smaller than CHF 500’000 to be acceptable given the size of the portfolio.
# Based on the result of my valuation, conclude whether the difference to the client’s estimate falls below his threshold. 
if absolute_difference < 500000:
    final_result = "It's to be acceptable given the size of the portfolio."
else:
    final_result = "It's to be unacceptable given the size of the portfolio."

print(final_result)
```
